```python
import os
import sys
parentdir = os.path.dirname(os.getcwd())
sys.path.append(parentdir)
from inquiryProcessor.dataset import InquiryDataset
from inquiryProcessor.inquiryEstimator import InquiryAnalyzerAssistant
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "../inquiryProcessor/inquiries_dataset.csv"
```

```python
dataset = InquiryDataset.get_training_dataset(path)
```

```python
dict = {}
for i in dataset:
    s = InquiryAnalyzerAssistant.classifierstring(i[1])
    if s in dict:
        dict[s].append(i[0])
    else:
        dict[s] = [i[0]]
```

```python
x = [len(i) for i in list(dict.values())]
labels = dict.keys()
map = plt.get_cmap("Greys", 85)
colors = [map(i) for i in x]
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
fig.set_dpi(80)
def absolute_value(val):
    a  = round(val/100*dataset.shape[0])
    return a
ax.pie(x, radius=20, colors=colors, center=(6,6), wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True, labels=labels, autopct=absolute_value, textprops={'color':'#8A806F'})
plt.show()
```
